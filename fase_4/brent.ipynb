```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
tables = pd.read_html('http://www.ipeadata.gov.br/ExibeSerie.aspx?module=m&serid=1650971490&oper=view', encoding='utf8')
```

```python
df = tables[2]
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
df.reset_index(inplace=True)
del df['index']
```

```python
df = df.rename(columns={'Data': 'data', 'Preço - petróleo bruto - Brent (FOB)': 'preco'})
```

```python
df.info()
```

```python
df.preco = df.preco.astype(float)
```

```python
df.preco = df.preco / 100
```

```python
df['data'] = pd.to_datetime(df['data'])
df
```

```python
df.info()
```

```python
df.preco.describe()
```

### Preço Petróleo bruto

```python
plt.figure(figsize=(13,6))
plt.plot(df.data, df.preco, color='red', linewidth=2)
plt.xlabel('Data')
plt.ylabel('Preço')
plt.title('Preço - petróleo bruto - Brent (FOB)')

# Add the grid
plt.grid(True)

# Show the plot
plt.show()
```

The document examines the causes and economic effects of the 2007–08 oil price shock compared to previous oil crises. Key takeaways:

1. **Causes of 2007–08 Oil Price Shock**:
   - Unlike earlier crises caused by supply disruptions, this shock stemmed from strong global demand (notably from China) and stagnant production.
   - Saudi Arabia's reduced output and declining production in regions like the North Sea exacerbated the issue.
   - Limited price elasticity of oil demand played a significant role in driving up prices.

2. **Economic Impact**:
   - The shock heavily affected consumption patterns, especially durable goods like automobiles.
   - The U.S. likely would not have been in a recession (2007Q4–2008Q3) without the consumption declines induced by high oil prices.

3. **Role of Speculation**:
   - Speculation in oil futures markets may have amplified price volatility but was not the primary cause.
   - Low price elasticity and stagnant production were more critical factors.

4. **Historical Context**:
   - Past oil shocks, usually triggered by geopolitical events, resulted in similar economic consequences, underscoring the strong link between oil price spikes and recessions.

5. **Policy Implications**:
   - Policymakers should consider the interplay between demand growth, production capacity, and market dynamics when addressing oil price volatility.

The sharp decline in oil prices after their peak in mid-2008 can be attributed to several interconnected factors:

### 1. **Global Economic Collapse**:
   - The financial crisis of 2008 caused a sudden and significant drop in global economic activity, reducing demand for oil. The sharp contraction in industrial production, transportation, and consumer spending led to a drastic reduction in oil consumption.

### 2. **Elasticity Effects**:
   - Short-term price elasticity of oil demand is low, meaning consumers and businesses initially find it difficult to reduce oil consumption in response to price increases. However, over time, they adapt by adopting more energy-efficient practices or cutting back on energy-intensive activities.
   - As these adjustments took effect, demand declined faster, amplifying the price drop.

### 3. **Speculative Dynamics**:
   - Speculative investment in oil futures may have inflated prices during the rise. When the economic downturn became evident, speculative positions were unwound rapidly, exacerbating the price collapse.

### 4. **Inventory Adjustments**:
   - During the price increase, inventories of oil were drawn down as refiners and consumers anticipated higher future prices. However, when prices started falling, inventories began to build up, putting further downward pressure on prices.

### 5. **Shifts in Perceived Demand**:
   - Market participants had underestimated the depth of the economic contraction. Once the extent of the downturn became clear, expectations about future oil demand were revised downward significantly, contributing to the price collapse.

### 6. **Lack of Immediate Supply Cuts**:
   - Oil producers, especially OPEC, did not immediately adjust production levels to the falling demand. The oversupply in the market accelerated the price decline.

In essence, the combination of a global financial crisis, delayed adjustments to demand and supply, and speculative unwinding created a perfect storm for a rapid and severe decline in oil prices.
