The dataset covers a wide range of information that is crucial for understanding the various factors associated with Alzheimer's disease. <li><strong>Physical Activity</strong>: Hours of physical activity per week are recorded, with values ranging from 0 to 10, to examine the relationship between exercise and cognitive function.</li>

<li><strong>Diabetes</strong>: Whether the patient has diabetes is included, since diabetes can increase the risk of developing Alzheimer's.</li>
<li><strong>Depression</strong>: The dataset records whether the patient has a history of depression, which is a known risk factor for cognitive decline.</li>
<li><strong>Head Injury</strong>: Information on whether the patient has suffered a head injury is included, as traumatic brain injuries are linked to an increased risk of Alzheimer's.</li>

<li><p><strong>Clinical Measurements</strong>:</p>

```python
# Importing libs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.datasets import make_classification
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
import joblib
```

```python
# Reading the dataa
data = pd.read_csv("alzheimers_disease_patient_data.csv", sep=",")
```

```python
data
```

```python
# Setting the index
data.set_index("PatientID", inplace=True)
```

```python
# deleting Doctor in Charge due to it is not needed
del data['DoctorInCharge']
```

```python
# Taking a glance on types and nan values
data.info()
```

```python
# Looking on initial stats
data.describe()
```

```python
# Looking to the correlation
correlation_matrix = data.corr().round(2)

fig, ax = plt.subplots(figsize=(20,20))    
sns.heatmap(data=correlation_matrix, annot=True, linewidths=.5, ax=ax)
```

```python
y = data['Diagnosis']
X = data[[col for col in data.columns if col != 'Diagnosis']]
X,y
```

```python
# using the train test split function 
X_train, X_test, y_train, y_test = train_test_split(X,y , 
                                   random_state=104,  
                                   test_size=0.25,  
                                   shuffle=True) 
```

```python
# Creating the model
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
```

```python
# Looking at the accuracy
clf.score(X_test, y_test)
```

```python
error = []

# Calculating error for max_depth values between 1 and 10
for i in range(1, 10):
    clf = RandomForestClassifier(max_depth=i, random_state=0)
    clf.fit(X_train, y_train) 
    pred_i = clf.predict(X_test)
    error.append(np.mean(pred_i != y_test)) 
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, 10), error, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate Max Depth Value')
plt.xlabel('N Value')
plt.ylabel('Mean Error')
```

```python
accuracy = []

# Calculating accuracy for max_depth values between 1 and 10
for i in range(1, 10):
    clf = RandomForestClassifier(max_depth=i, random_state=0)
    clf.fit(X_train, y_train) 
    pred_i = clf.score(X_test, y_test)
    accuracy.append(pred_i)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(1, 10), accuracy, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('Accuracy Rate Max Depth Value')
plt.xlabel('N Value')
plt.ylabel('Accuracy')
```

As we can see the N value for max depth which fits better is 8 or 9, because the accurary is high and the error low

```python
# Creating the model with max depth = 9
clf = RandomForestClassifier(max_depth=9, random_state=0)
clf.fit(X_train, y_train)
```

```python
# Looking at the accuracy
clf.score(X_test, y_test)
```

```python
y_pred = clf.predict(X_test)
y_pred
```

```python
# Now let's analyze the confusion matrix
y_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, 
            annot=True,
            fmt='g', 
            xticklabels=['Alzheimer','Not Alzheimer'],
            yticklabels=['Alzheimer','Not Alzheimer'])
plt.ylabel('Actual', fontsize=13)
plt.title('Confusion Matrix', fontsize=17, pad=20)
plt.gca().xaxis.set_label_position('top') 
plt.xlabel('Prediction', fontsize=13)
plt.gca().xaxis.tick_top()

plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, 'Prediction', ha='center', fontsize=13)
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```
